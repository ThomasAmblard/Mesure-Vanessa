```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

TABLE = pd.read_csv('Classeur1.csv',sep=';')

display(TABLE)
```

```python
# je remplace les legendes par les modalités pour la colonne sex ( 0 = garcon et 1= filles)
TABLE['Sex'] = TABLE['Sex'].replace([0], 'Garçon')
TABLE['Sex'] = TABLE['Sex'].replace([1], 'Fille')

# je remplace les legendes par les modalités pour la colonne Ethnie 
TABLE['Ethnie'] = TABLE['Ethnie'].replace([1], 'Caucasien')
TABLE['Ethnie'] = TABLE['Ethnie'].replace([2], 'Noir')
TABLE['Ethnie'] = TABLE['Ethnie'].replace([3], 'Asiatique')
TABLE['Ethnie'] = TABLE['Ethnie'].replace([4], 'Magrébin')

# je remplace les legendes par les modalités pour la colonne Facial type. 
TABLE['Facial Type'] = TABLE['Facial Type'].replace([1], 'Leptoprosopic >104')
TABLE['Facial Type'] = TABLE['Facial Type'].replace([2], 'Mesoprosopic 97-104')
TABLE['Facial Type'] = TABLE['Facial Type'].replace([3], 'Euryprosopic <97')

# je remplace les legendes par les modalités pour la colonne ODF ( traitement ortho.). 
TABLE['ODF'] = TABLE['ODF'].replace([0], 'Pas de traitement')
TABLE['ODF'] = TABLE['ODF'].replace([1], 'traitement')
TABLE['ODF'] = TABLE['ODF'].replace([1.1], "traitement 'Orthopedic ttt'")
TABLE['ODF'] = TABLE['ODF'].replace([1.2], "traitement 'Orthodontic TTT'")
TABLE['ODF'] = TABLE['ODF'].replace([1.3], "traitement 'Surgical TTT'")
```

## <span style='font-family:Garamond'><span style='color:#046307'> Information age pour l'ensemble de la population et par sex.
</span>

```python

Information_Age = TABLE['Age'].describe()
info_age_fille = TABLE['Age'].loc[TABLE['Sex']=='Fille'].describe()
info_age_garcon = TABLE['Age'].loc[TABLE['Sex']=='Garçon'].describe()

print('Info age general')
print(Information_Age)
print()
print('Info age garcon')
print(info_age_garcon)
print()
print('Info age fille')
print(info_age_fille)

Information_Sex = TABLE['Sex'].value_counts(normalize=True) * 100
TABLE["Sex"].value_counts(normalize=True).plot(kind='pie',autopct='%1.0f%%',shadow =True,title='Proportions Homme et Femme',figsize=(20,10))
print('Proportions sex')
display(Information_Sex)
```

## <span style='font-family:Garamond'><span style='color:#046307'> Effectifs et proportions de chaque type facial pour la population total, les hommes et les femmes.</span>

```python
#Creation de la table effectifs approrié.
EFFECTIFS_TYPE_FACIAL = TABLE.groupby(['Sex', 'Facial Type'])['Facial Type'].count().to_frame()
EFFECTIFS_TYPE_FACIAL = EFFECTIFS_TYPE_FACIAL.rename(columns={'Facial Type': 'Effectif'})
EFFECTIFS_TYPE_FACIAL = EFFECTIFS_TYPE_FACIAL.pivot_table(EFFECTIFS_TYPE_FACIAL, index=["Sex"],columns=["Facial Type"],aggfunc=["mean"]).reset_index()
EFFECTIFS_TYPE_FACIAL.loc[2]=['Garçons + Filles', 34+18 , 55+49, 92+52 ]
EFFECTIFS_TYPE_FACIAL['Sex'] = EFFECTIFS_TYPE_FACIAL['Sex'].replace(to_replace=[0,1],value=['Garcons','Filles'])
EFFECTIFS_TYPE_FACIAL = EFFECTIFS_TYPE_FACIAL.set_index('Sex')
display(EFFECTIFS_TYPE_FACIAL)

#Creation de la table proportions approrié.
PROPORTIONS_TYPE_FACIAL = pd.DataFrame(
{
'Sex' : ['Fille','Garçon','Garçons et Fille'], 
'Euryprosopic <97' : [18.79,15.13,17.33], 
'Leptoprosopic >104' : [30.39, 41.18,34.67],
'Mesoprosopic 97-104' : [50.82, 43.70,48 ]  
}
)
PROPORTIONS_TYPE_FACIAL = PROPORTIONS_TYPE_FACIAL.set_index('Sex')

display(PROPORTIONS_TYPE_FACIAL)

# Graphiques
EFFECTIFS_TYPE_FACIAL.plot.bar(figsize=(15,10),title='Effectifs type facial')
plt.ylabel('Effectif')
PROPORTIONS_TYPE_FACIAL.plot.bar(figsize=(15,10),title='proportions type facial')
plt.ylabel('%')

```


## <span style='font-family:Garamond'><span style='color:#046307'> Nose to chin par type facial et sex.</span>

```python
#Creation de la table approrié.
NOSE_TO_CHIN_DISTANCE_SEX = TABLE.groupby(['Sex', 'Facial Type'])['Nose-to-Chin Distance'].mean().to_frame()
NOSE_TO_CHIN_DISTANCE_SEX = NOSE_TO_CHIN_DISTANCE_SEX.pivot_table(NOSE_TO_CHIN_DISTANCE_SEX, index=["Sex"],columns=["Facial Type"],aggfunc=["mean"]).reset_index()
NOSE_TO_CHIN_DISTANCE_SEX.loc[3]=['Fille + Garçon', 68.77 , 76.255, 72.4845315 ]
NOSE_TO_CHIN_DISTANCE_SEX = NOSE_TO_CHIN_DISTANCE_SEX.set_index('Sex')
display(NOSE_TO_CHIN_DISTANCE_SEX)

# Graphique 
NOSE_TO_CHIN_DISTANCE_SEX.plot.bar(figsize=(15,10))
plt.ylabel('Distance')
```

## <span style='font-family:Garamond'><span style='color:#046307'> Nose to chin pour personne ayant eu un traitement ou non.</span>

```python
#Creation de la table approrié.
NOSE_TO_CHIN_DISTANCE_TRAITEMENT = TABLE.groupby(['ODF', 'Facial Type'])['Nose-to-Chin Distance'].mean().to_frame()
NOSE_TO_CHIN_DISTANCE_TRAITEMENT = NOSE_TO_CHIN_DISTANCE_TRAITEMENT.pivot_table(NOSE_TO_CHIN_DISTANCE_TRAITEMENT, index=["ODF"],columns=["Facial Type"],aggfunc=["mean"]).reset_index()
NOSE_TO_CHIN_DISTANCE_TRAITEMENT = NOSE_TO_CHIN_DISTANCE_TRAITEMENT.set_index('ODF')
display(NOSE_TO_CHIN_DISTANCE_TRAITEMENT)

# Graphique 
NOSE_TO_CHIN_DISTANCE_TRAITEMENT.plot.bar(figsize=(15,10))
plt.ylabel('Distance')
```
